**MODULAR EXPONENTIAL METHOD**

```python
def fme(a,d,n):
  ans=1;
  while d!=0:
    if d%2==1:
      ans=((ans%n)*(a%n))%n
    a=((a%n)*(a%n))%n
    d>>=1
  return ans;
for i in range(len(data)):
  start=time.time() 
  answ=fme(data[i][0],data[i][1],data[i][2])
  execution=time.time()-start
  exec_e.append(execution)
  a_1.append(answ)
```

**GENERAL METHOD**

```python
exec_g=[]
a_2=[]
def g_m(a,d,n):
  ans=0
  ans=(a**d)%n
  return ans
for i in range(len(data[:7])):
  start_a=time.time() 
  an=g_m(data[i][0],data[i][1],data[i][2])
  execution_a=time.time()-start_a
  exec_g.append(execution_a)
  a_2.append(an)
```

**EFFICIENCY**

```python
eff=[]
k=0
for i in range(len(data[:7])):
  k=((exec_g[i]-exec_e[i])/(exec_g[i]))*100
  eff.append(k)
e=sum(eff)/len(eff)
print('The Average Efficiency of Fast Modular Exponentation Over General Modular Multiplication = {}'.format(e))
```
