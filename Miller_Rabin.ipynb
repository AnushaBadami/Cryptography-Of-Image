```python
mi_time =[]
isprime_mi = []

def is_prime_miller(n, k=40):
    if n == 2 or n == 3:
        return True
    if n <= 1 or n % 2 == 0:
        return False
    s = 0
    r = n - 1
    while r & 1 == 0:
        s += 1
        r //= 2
    for _ in range(k):
        a = randrange(2, n - 1)
        x = pow(a, r, n)
        if x != 1 and x != n - 1:
            j = 1
            while j < s and x != n - 1:
                x = pow(x, 2, n)
                if x == 1:
                    return False
                j += 1
            if x != n - 1:
                return False
    return True

for i in range(l):
   start = time.time()
   correct = is_prime_miller(prime[i])
   isprime_mi.append(correct)
   end = time.time()-start 
   mi_time.append(end)
print(isprime_mi)
print(mi_time)
```

```python
gen_time=[]
isprime_gen=[]

def is_prime_gen(num):
  flag=False
  for i in range(2, (num//2)+1):
    if (num % i) == 0:
      flag = True
      break
  if flag:
      return False
  else:
      return True

for i in range(17):
   start = time.time()
   correct = is_prime_gen(prime[i])
   isprime_gen.append(correct)
   end = time.time()-start 
   gen_time.append(end)
print(isprime_gen)
print(gen_time)
```

```python
plt.plot(n_bit[:17],gen_time[:17],label='Original Method')
plt.plot(n_bit[:17], mi_time[:17],label='Miller Rabbin Algorithm')
plt.scatter(n_bit[:17], gen_time[:17])
plt.scatter(n_bit[:17], mi_time[:17])
plt.xlabel('Number of Bits')
plt.ylabel('Execution time')
plt.title('Time Comaparion')
plt.legend()
plt.show()
```

```python
eff=[]
k=0
for i in range(17):
  k=((gen_time[i]-mi_time[i])/(gen_time[i]))*100
  eff.append(k)
e=sum(eff)/len(eff)
print('The Average Efficiency Calculated For Random Prime Number : {}'.format(e))
```
