**KARATSUBA MULTIPLICATION METHOD**

```python
def karatsuba(x,y):
  if len(str(x)) < 3 or len(str(y)) < 3:
      return x*y

  n = max(len(str(x)),len(str(y))) // 2

  a = x // 10**(n)
  b = x % 10**(n)
  c = y // 10**(n)
  d = y % 10**(n)

  z = karatsuba(b,d)
  z1 = karatsuba((a+b), (c+d))
  z2 = karatsuba(a,c)
  answer=((10**(2*n))*z2)+((10**n)*(z1-z2-z))+z
  return answer
```

```python
def general_multiplication(X, Y):
    x = str(X)
    y = str(Y)
    answer = 0
    for i in range(len(y)):
        c = 0 
        wr = "" 
        for j in range(len(x) - 1, -1, -1):
            n = int(y[i]) * int(x[j]) + c
            if n > 9 and j > 0:
                wr = str(n % 10) + wr
                c = n // 10
            else:
                wr = str(n) + wr
                c = 0
        answer*= 10
        answer += int(wr)
    return answer
```
